```python
ratings = pd.read_csv('ratings1.csv')
ratings.head()  
```

```python
ratedMovies = ratings.copy()
```

```python
aboverating = ratedMovies[ (ratedMovies['movieId'] > 4806)].index
```

```python
ratedMovies.drop(aboverating, inplace=True, axis=0 )
```

```python
print('The dataset contains: ', len(movies))
```

```python
def getGenreRatings(ratings, movies, genres, column_names):
    genreRatings = pd.DataFrame()
    for genre in genres:
        genreMovie = movies[movies['genres'].str.contains(genre)]
        avgGenreVotesPerUser = ratings[ratings['movieId'].isin(genreMovie['movieId'])].loc[:, ['userId','rating']].groupby(['userId'])['rating'].mean().round(2)
        genreRatings = pd.concat([genreRatings, avgGenreVotesPerUser], axis = 1)
    genreRatings.columns = column_names
    return genreRatings
```

```python
genreRatings = getGenreRatings(ratings,movies, ['Action', 'Romance'],['avgActionRating', 'avgRomanceRating'])
genreRatings.head()
```
